#### –ú–æ–∏ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è –¥–∞–ª—å–Ω–µ–π—à–µ–π —Ä–∞–±–æ—Ç—ã

```python
def iqr(df, column):
    '''
    –ù–∞—Ö–æ–¥–∏–º –≥—Ä–∞–Ω–∏—Ü—ã –¥–ª—è –∞—É—Ç–ª–∞–µ—Ä–æ–≤
    '''
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    quartile_lower = Q1 - 1.5 * IQR
    quartile_higher = Q1 + 1.5 * IQR
    return quartile_lower, quartile_higher

def cat_feat_plot(df, x_list, y_name):
    '''
    –ì—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö
    '''
    for el in x_list:
        plt.figure(figsize=(10, 5))
        plt.subplot(1,2,1)
        plt.title(el + ' histogram')
        sns.countplot(data=df, x=el)

        plt.subplot(1,2,2)
        plt.title(el + ' pointplot')
        sns.pointplot(data=df, x=el, y=y_name)

        plt.show()
        
def num_feat_plot(df, x_list, y_name):
    '''
    –ì—Ä–∞—Ñ–∏–∫–∏ –¥–ª—è —á–∏—Å–ª–æ–≤—ã—Ö –ø–µ—Ä–µ–º–µ–Ω–Ω—ã—Ö
    '''
    for el in x_list:
        plt.figure(figsize=(10, 5))
        plt.subplot(1,2,1)
        plt.title(el + ' histogram')
        sns.histplot(data=df, x=el) # hue=y_name —É–±—Ä–∞–ª–∞

        plt.subplot(1,2,2)
        plt.title(el + ' boxplot')
        sns.boxplot(data=df, x=el)

        plt.show()
```

### –ú–æ–¥–µ–ª–∏—Ä–æ–≤–∞–Ω–∏–µ

–ò—Å–ø–æ–ª—å–∑—É–µ–º –ª–∏–Ω–µ–π–Ω—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏—é —Å L1 —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–µ–π, —á—Ç–æ–±—ã –æ—Ç–æ–±—Ä–∞—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–∏. –ò—Å–ø–æ–ª—å–∑—É–µ–º –ø–æ–∫–∞ –≤—Å–µ –¥–∞–Ω–Ω—ã–µ –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏.

```python
train_df = data[f_bin + f_num_mm + f_num_ss + ['SalePrice']].copy()
X = train_df.drop(['SalePrice'], axis=1).copy()
y = train_df['SalePrice'].copy()
```

```python
standardscaler = StandardScaler()
standardscaler.fit(X[f_num_ss])
X[f_num_ss] = standardscaler.transform(X[f_num_ss])

minmax = MinMaxScaler()
minmax.fit(X[f_num_mm])
X[f_num_mm] = minmax.transform(X[f_num_mm])
```

```python
# –æ–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å –∏ –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ–º
estimator = Lasso(max_iter=20000)
estimator.fit(X, y)
y_pred = estimator.predict(X)

# –æ—Ü–µ–Ω–∏–≤–∞–µ–º –∫–∞—á–µ—Å—Ç–≤–æ –º–æ–¥–µ–ª–∏
print(f'MAE: {mean_absolute_error(y, y_pred)}')
print(f'MSE: {mean_squared_error(y, y_pred)}')
print(f'RMSE: {mean_squared_error(y, y_pred) ** 0.5}')
```

```python
coef_dict = {}
for coef, feat in zip(estimator.coef_, list(train_df.columns)):
    coef_dict[feat] = coef

result = pd.DataFrame.from_dict(coef_dict, orient='index', columns=['Coef'])
```

```python
f_bin_new = []
for el in f_bin:
    if el not in list(result[result['Coef'] == 0].index):
        f_bin_new.append(el)
```
